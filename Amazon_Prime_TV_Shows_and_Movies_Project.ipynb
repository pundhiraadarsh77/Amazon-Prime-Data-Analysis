## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
````

````python
# First of all, dropping duplicate values present in the dataset.
data = data.drop_duplicates()
````

````python
# Now handling null values for each column
````

````python
# 1. For "description" column - 0.15% null values present

'''
Filling null values with "Unknown" as the data type of "description" column is object.
So filling null with "Unknown" is a better approach than dropping those records.
'''

data["description"] = data["description"].fillna("Unknown")
````

````python
# 2. For "age_certification" column - 54.58% null values present

'''
Filling null values of "age_certification" column with "Unknown".
This column data type is also object and there is no other way to determine the column
so filling it with "Unknown".
'''

data["age_certification"] = data["age_certification"].fillna("Unknown")
````

````python
# For "genres" column - genres as "[]" considered as null values

# First replacing those values with null so we can handle them
data.loc[data["genres"] == "[]", "genres"] = np.nan

'''
Filling null values of "genres" column with "['unknown']".
This column data type is also object and there is no other way to determine the column
so filling it with "['unknown']".
'''

data["genres"] = data["genres"].fillna("['unknown']")
````

````python
# For "production_countries" column - production_countries as "[]" considered as null values

# First replacing those values with null so we can handle them
data.loc[data["production_countries"] == "[]", "production_countries"] = np.nan

'''
Filling null values of "production_countries" column with "['unknown']".
This column data type is also object and there is no other way to determine the column
so filling it with "['unknown']".
'''

data["production_countries"] = data["production_countries"].fillna("['unknown']")
````

````python
# 3. For "seasons" column - 93.20% null values present

'''
The records where seasons column is null are for MOVIE and not for TV SHOW.
A movie cannot have season that's why we are filling the null values of this column as 0.
'''

data["seasons"] = data["seasons"].fillna(0)
````

````python
# 4. For "imdb_id" column - 4.29% null values present

'''
We cannot fill null values with random unique id for this column.
So again, we are filling them with "Unknown" which also do not make any effect
on our analysis as the data type of column is object.
'''

data["imdb_id"] = data["imdb_id"].fillna("Unknown")
````

````python
# Handling null values of "imdb_score" and "imdb_votes"

# 5. For "imdb_score" column - 5.08% null values present

# 6. For "imdb_votes" column - 5.10% null values present

'''
Filling both the column's null values with median value of respective column.

Why median and not mean?
- Because mean can get distorted by huge values present in the dataset but
  median stays stable in those cases.
'''

data["imdb_score"] = data["imdb_score"].fillna(data["imdb_score"].median())
data["imdb_votes"] = data["imdb_votes"].fillna(data["imdb_votes"].median())
````

````python
# Handling null values of "tmdb_popularity" and "tmdb_score"

# 7. For "tmdb_popularity" column - 0.44% null values present

# 8. For "tmdb_score" column - 8.78% null values present

'''
As we know, mean gets distorted by huge values and median stays stable.
We are going to fill the null values of these columns with median.
'''

data["tmdb_popularity"] = data["tmdb_popularity"].fillna(data["tmdb_popularity"].median())
data["tmdb_score"] = data["tmdb_score"].fillna(data["tmdb_score"].median())
````

````python
# 9. Handling null values of "person_id", "name" and "role" columns

'''
All these 3 columns have equal null values present - 0.80%.
So if we drop the null values of any of the column, the other two columns will automatically gets sorted
and will not contain null values.
As the null values percent is very small(0.8%) we are dropping the null values of "person_id" column.
'''

data = data.dropna(subset= ["person_id"])
````

````python
# 10. For "character" column - 13.80% null values present

'''
The data type of this column is object.
So filling null values with "Unknown" for this column will be appropriate.
'''

data.loc[:, "character"] = data["character"].fillna("Unknown")
````

````python
# Rechecking null values present in the dataset
print(data.isnull().sum())
````

````python
# Now fixing the data types of column where needed

# For "seasons" column - Converting the data type of this column from float to integer as seasons are naturally integers and can not be fractional
data["seasons"] = data["seasons"].astype(int)

# For "imdb_votes" column - Converting the data type of this column from float to integer as votes are naturally discrete and can not be fractional
data["imdb_votes"] = data["imdb_votes"].astype(int)

# For "tmdb_popularity" column- No need for values upto 3 decimals, so rounding values to 2 decimals
data["tmdb_popularity"] = data["tmdb_popularity"].round(2)

# For "person_id" column - As this column represents unique id for person, it can't be in decimals so converting it into integer
data["person_id"] = data["person_id"].astype(int)
````

````python
# Checking if any numerical column contains negative value or not

# Selecting columns with data type as either integer or float
cols = data.select_dtypes(include= ["int64", "float64"])

# Checking for any negative value
(cols<0).sum()
````

````python
# Checking for any outliers present in the dataset

data.describe(include= ["int64", "float64"]).transpose()
````

````python
'''
On analysing numerical columns, some movie's runtime values are extremely less.
So we are going to manually filling them with their respected runtimes.
'''

data.loc[data["title"] == "Incognito", "runtime"] = 108
data.loc[data["title"] == "Evangelion: 3.0+1.0 Thrice Upon a Time", "runtime"] = 155
````

````python
# Removing unnecessary symbols in column "genres" to improve readability

data["genres"] = data["genres"].str.replace("[\[\]']", "", regex = True)
````

````python
# Removing unnecessary symbols in column "production_countries" to improve readability

data["production_countries"] = data["production_countries"].str.replace("[\[\]']", "", regex = True)
````

````python
# Creating a runtime_category column which returns the value as short, standard, long based on runtimes for movies/ tv shows.

data["runtime_category"] = pd.cut(
    data["runtime"],
    bins= [0,60,120,180,600],
    labels= ["Short","Standard","Long","Very Long"]
)
````

````python
# Creating a rating_category column which returns the value as low, average, high based on imdb_scores for movies/ tv shows.

data["rating_category"] = pd.cut(
    data["imdb_score"],
    bins= [0,4,6,8,10],
    labels= ["Poor","Average","Good","Excellent"]
)
````

````python
'''
After merging the title and cast datasets each title appeared multiple times due to many cast members.
Thus this can affect title-level analysis so we are creating a separate dataset which contains unique titles and runtime.
So for actor-level analysis we will use combined dataset, and for title-level analysis we are going to use dataset containing new titles.
'''

movie_data = data.drop_duplicates(subset= ["title", "runtime"])
````

````python
# Aggregating movie_data by type to calculate no. of unique titles.
# Using .nunique() function to calculate unique no. of titles per type.

movie_data.groupby("type")["title"].nunique().reset_index(name = "Total titles")
````

````python
# Aggregating data by release_year to count No. of Movie or TV shows released in that particular year.


movie_data.groupby("release_year")["title"].count().sort_values(ascending= False).reset_index(name = "no. of titles")
````

````python
# Aggregating data by type to find average imdb_score achieved by different type i.e Movie or Tv show.

data.groupby("type").agg(average_rating = ("imdb_score", "mean")).reset_index()
````

````python
# Aggregating data by role to find total no. of actors and directors in the dataset.

data.groupby("role")["title"].nunique().reset_index(name= "count")
````

````python
# Creating actor dataset to find actors appearing in most movies
actors = data[data["role"] == "ACTOR"]

# Aggregating actors data by name to find top 10 actors appearing in multiple movie/ shows.
actors.groupby("name")["title"].count().sort_values(ascending= False).reset_index(name = "movie/show count").head(10)
````

````python
# Creating director dataset to find director with most movies
director = data[data["role"] == "DIRECTOR"]

# Aggregating actors data by name to find top 10 directors with most movie/ shows.
director.groupby("name")["title"].count().sort_values(ascending= False).reset_index(name = "movie/show count").head(10)
````

````python
# Finding top 10 genres having versatility and can blend with other genres easily.

movie_data["genres"].value_counts().reset_index(name= "title_count").sort_values(by= "title_count", ascending= False).head(10)
````

````python
# Finding top 10 genres liked by audience most.

# Creating genres dataframe to analyze most popular genres.
genre_df = movie_data.assign(genres = movie_data["genres"].str.split(", ")).explode("genres")

genre_df["genres"].value_counts().reset_index(name= "title_count").sort_values(by= "title_count", ascending= False).head(10)
````

````python
# Finding top 10 production_countries.

# Creating production_countries dataframe to analyze production_countries.
country_df = movie_data.assign(production_countries = movie_data["production_countries"].str.split(", ")).explode("production_countries")

country_df["production_countries"].value_counts().reset_index(name= "title_count").sort_values(by= "title_count", ascending= False).head(10)
````

````python
# Aggregating data by type to calculate average runtime for Movies and Tv shows

data.groupby("type").agg(average_runtime = ("runtime", "mean")).reset_index()
````

````python
# Observing relationship between the two ratings - imdb_score and tmdb_score, to understand their evaluating process.

movie_data[["imdb_score","tmdb_score"]].corr()
````

````python
# Calculating the no. of titles across different runtime categories.

data["runtime_category"].value_counts().reset_index(name= "no. of titles")
````

````python
# Calculating the distribution of "Excellent" rated titles by runtime category.

data.loc[data["rating_category"]=="Excellent"]["runtime_category"].value_counts().reset_index(name= "title_count")
````

````python
# Analysing the distribution of titles having different ratings across different runtime categories.

ct = pd.crosstab(data["type"], data["rating_category"])
print(ct)

'''
On analysing this, it shows that our dataset has no. of movies more than shows.
Also it shows, poor-rated movies(8,880 titles) are way more than the poor-rated shows(95 titles) which suggests although movie are numerous,
show tends to maintain strong audience reception with relatively fewer poorly rated titles.
'''
````
